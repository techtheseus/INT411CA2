```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the different estimation techniques
technique_names = ['Expert Judgment', 'Algorithmic Models', 'Machine Learning']
mae_values = [2.3, 3.1, 1.8]
rmse_values = [3.5, 4.2, 2.1]
r_squared_values = [0.8, 0.6, 0.9]
bias_values = [-0.2, 0.5, -0.1]

# Create a scatter plot for the MAE values
plt.scatter(technique_names, mae_values)
plt.xlabel('Estimation Technique')
plt.ylabel('MAE')
plt.title('Comparison of Estimation Techniques using MAE')
plt.show()

# Create a scatter plot for the RMSE values
plt.scatter(technique_names, rmse_values)
plt.xlabel('Estimation Technique')
plt.ylabel('RMSE')
plt.title('Comparison of Estimation Techniques using RMSE')
plt.show()

# Create a scatter plot for the R values
plt.scatter(technique_names, r_squared_values)
plt.xlabel('Estimation Technique')
plt.ylabel('R')
plt.title('Comparison of Estimation Techniques using R')
plt.show()

# Create a scatter plot for the Bias values
plt.scatter(technique_names, bias_values)
plt.xlabel('Estimation Technique')
plt.ylabel('Bias')
plt.title('Comparison of Estimation Techniques using Bias')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
techniques = ['Expert judgment', 'Algorithmic models', 'Machine learning']
mae_values = [2.5, 3.2, 1.8]
rmse_values = [3.8, 4.5, 2.3]

# Create figure and axis objects
fig, ax = plt.subplots()

# Set bar width
bar_width = 0.35

# Set x locations for the bars
xlocs = np.arange(len(techniques))

# Create bar plots for MAE and RMSE
mae_bars = ax.bar(xlocs - bar_width/2, mae_values, width=bar_width, label='MAE')
rmse_bars = ax.bar(xlocs + bar_width/2, rmse_values, width=bar_width, label='RMSE')

# Set axis labels and legend
ax.set_xlabel('Effort Estimation Techniques')
ax.set_ylabel('Error (days)')
#ax.set_title('Comparison of MAE and RMSE values for different techniques')
ax.set_xticks(xlocs)
ax.set_xticklabels(techniques)
ax.legend()

# Add value labels to the bars
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom')

autolabel(mae_bars)
autolabel(rmse_bars)

# Show the plot
plt.show()
```
